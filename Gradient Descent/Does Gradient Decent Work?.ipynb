```python
import numpy as np
import matplotlib.pyplot as plt
import gradient_descent as gd
%matplotlib inline
```

```python
def get_parameters():
    m = np.random.randint(100,250)
    n = 1 #np.random.randint(1,5)
    # y = ax + b
    a = np.random.randint(4,10)
    b = np.random.randint(2,4)
    x = np.random.rand(m,n)
    y = a*x + b + np.random.random((m,1))
    return x,y,a,b
```

```python
def plot(x1,y1):
    plt.plot(x, y, 'b.')
    plt.xlabel('x')
    plt.ylabel('y')
    
def plotAll(x,y,points,y_predict,total_iterations,cost_history):
    fig, ax = plt.subplots(1,2, figsize=(12,5))
    fig.tight_layout(pad=5.0)
    
    ax[0].plot(x, y, 'b.')
    ax[0].plot(points, y_predict, 'r-')
    ax[0].set_xlabel('x')
    ax[0].set_ylabel('y')
    
    ax[1].plot(range(total_iterations), cost_history, 'g.')
    ax[1].set_ylabel('cost')
    ax[1].set_xlabel('iterations')
```

```python
x,y,a,b = get_parameters()
total_iterations = 10000
print('Line is %d*x + %d' % (a, b))
gdescent = gd.GradientDescent()
points = np.array( [[0],[1.0]] )
plot(x,y)
```

```python
theta, cost_history, theta_history = gdescent.gradient_descent(x,y, iterations=total_iterations, learning_rate=0.001)
y_predict = gdescent.predict(points, theta)
plotAll(x,y,points,y_predict,total_iterations,cost_history)
plotAll(x,y,points,y_predict,200,cost_history[0:200]) #to show that it is constantly decreasing
```

```python
total_iter = 5000
theta, cost_history, theta_history = gdescent.stochastic_gradient_descent(x,y, iterations=total_iter, learning_rate=0.001)
y_predict = gdescent.predict(points, theta)
plotAll(x,y,points,y_predict,total_iter,cost_history)
plotAll(x,y,points,y_predict,200,cost_history[0:200]) #to emphasize that for stochastic, error doesn't always decrease
```

```python
theta, cost_history, theta_history = gdescent.minibatch_gradient_descent(x,y, iterations=total_iterations, learning_rate=0.001)
y_predict = gdescent.predict(points, theta)
plotAll(x,y,points,y_predict,total_iterations,cost_history)
plotAll(x,y,points,y_predict,200,cost_history[0:200]) #doesn't oscillate as much, but does
```
