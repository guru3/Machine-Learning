We use gradient descent in machine learning to find minimas of our cost functions for our models. Idea being to move in direction opposite of the gradient of the function, to reach the minima in the function.


Now, let's try for a 2D function

```python
f = lambda x, y: (x-5)**2 + (y-5)**2 + 3
dfx = lambda x: 2*(x-5) #x gradient
dfy = lambda y: 2*(y-5) #y gradient

step_size = 1000 #random big number
iterCounter = 0
learning_rate = 0.01 # Learning rate
precision = 1e-6  #acceptable precision
max_iters = 10000 # max #iterations, in case we do not reach precision requirement


currPosX = 0.0;
currPosY = 0.0;
positionsAre = [ [currPosX, currPosY] ];
while step_size > precision and iterCounter < max_iters:
    lastPos = [currPosX, currPosY]
    currPosX = currPosX - learning_rate * dfx(lastPos[0]) #gradient descent step
    currPosY = currPosY - learning_rate * dfy(lastPos[1]) #gradient descent step
    
    positionsAre.append( [currPosX,currPosY] )
    step_size = np.sqrt((currPosX - lastPos[0])**2 + (currPosY - lastPos[1])**2)
    if iterCounter%100 == 0:
        print("Iteration",iterCounter,"\nValue is",[currPosX,currPosY]) #Print iterations
    iterCounter = iterCounter + 1
    
print("The local minimum occurs at", currPos)
```
